The Department of Finance (DOF) is required by NY State law to value condominiums or cooperatives as if they were residential 
rental apartment buildings. DOF uses income information from rental properties similar in physical features and location to 
the condominiums or cooperatives. DOF applies this income data to the condominium or cooperative and determine its value in 
the same way DOF values rental apartment buildings. Update Schedule: Annually

The overall goal of the project is to predict if there is a correlation between crime in a neighborhood and the full market value of the Condominiums. The analysis currently presented is to see if other predictors are able to predict the Condominium's **Full Market Value**, which is our **tarket variable**.

**Target Variable: Full Market Value**

The *full market value* is the value that represents the price a Condominium building would sell for.  However, there could be many
predictors that can influence this value, such as the neighborhood the Condominuim is located, number of units, age, estimated square
foot to name a few. Understanding the full market value can assist in estimating the price of new buildings of similar style, of if 
a building goes on the market estimating that buildings value given a predictor model.

**Features**:
    The features I have chosen to use as predictors for the full market value are:
* ** Gross Sqft ** - The sum of all areas on all floors of a building included within the outside faces of its exterior walls,     including all vertical penetration areas, for circulation and shaft areas that connect one floor to another.
* **Estimated Gross Income** - an individual's total gross income minus specific deductions
* **Gross Income per Sqft** - Sales per square foot is simply the average revenue a retail business creates for every square foot of sales space. 
* **Estimated Expense** - expenses that are not actual or real. it may be more than or less than the actual expenses
* **Expense per Sqft** - is a calculation of the value of each square foot of area of a house, condo or any building. It is a simple, but useful calculation that is mostly used to compare similar properties
* **Net Operating Income** - the annual income generated by an income-producing property after taking into account all income collected from operations, and deducting all expenses incurred from operations
* **Market Value per Sqft** - If you are comparing a subject home to homes that have sold in the area, then you can use the prices of the sold homes to calculate a price per square foot for each 

These features all take into account the cost of a condominum for year to year, which might give us an indication of the predicting
the full market value. Plus, these values should remain relatively constant from year to year unless something major happens in the
market.

The Model: Without the transformation of the Target Variable

```python
from sklearn import feature_selection as f_select
from sklearn import linear_model as lm
from sklearn import cross_validation as cv
from sklearn import metrics
```

```python
# Make sure the data is in a dataframe
df_Condo_predictor = pd.DataFrame(df_Condo_predictor)

# List all the columns in the dataframe
x_columns = list(df_Condo_predictor.columns)

# Explictly define the y_column
y_column = 'Full Market Value'

# Remove the y column from the list of columns
x_columns.remove(y_column)

# Define an empty list to put the relevant features that will predict the target variable
significant_columns = []

# Define another empty list to put the pvalues
pvals = []

# A small script that will go through each of the columns in the dataframe and run a regression analysis on the feature and the
# predictor data sets. It will calculate the p-values and a f-statistic, which is a comparison of this how this data fits vs 
# the same data set with less information. If the p values are below 0.05 then we will append that feature and its p-value.

for feature in x_columns:
    pval = f_select.f_regression(df_Condo_predictor[[feature]], df_Condo_predictor[y_column])
    if pval[1][0] < 0.05:
        significant_columns.append(feature)
        pvals.append(pval[1][0])
```

```python
# Now going to split the arrays and data frames into two unique and random sets.
x_train, x_test, y_train, y_test = cv.train_test_split(df_Condo_predictor[significant_columns], df_Condo_predictor[y_column],
                                                       test_size=0.333,random_state=1234)

#Fit the model using the linear regression model
model = lm.LinearRegression().fit(x_train, y_train)
```

```python
# Print out the results of the test
print pd.DataFrame({
    'column': significant_columns,
    'coef': model.coef_,
    'p-value': pvals,
}).set_index('column')
```

```python
# Compare the training data sets r^2 value for a group of features and the target value
print model.score(x_train, y_train)
print metrics.r2_score(y_train, model.predict(x_train))
```

```python
# Compare the test data sets r^2 value for a group of features and the target value
print model.score(x_test, y_test)
print metrics.r2_score(y_test, model.predict(x_test))
```
