The Department of Finance (DOF) is required by NY State law to value condominiums or cooperatives as if they were residential 
rental apartment buildings. DOF uses income information from rental properties similar in physical features and location to 
the condominiums or cooperatives. DOF applies this income data to the condominium or cooperative and determine its value in 
the same way DOF values rental apartment buildings. Update Schedule: Annually

```python
import statsmodels.formula.api as smf
# OLS, or ordinary least squares, takes a y (dependent variable) and X (independent variables) (formula = y ~ X)
# Below, we copy the data frame and remove the na variables, and create a single variable linear model
# to return a test statistic and p-value, to see how strong of a relationship bodyweight and brainweight have.
#fig, axes = plt.subplots(nrows=1,ncols=2)

df_linear = df_Condo1[['Full Market Value', 'Estimated Gross Income']]
df_linear.columns = ['Full_Market_Value', 'Estimated_Gross_Income']
model = smf.ols(formula='Full_Market_Value ~ Estimated_Gross_Income' , data=df_linear)
results = model.fit()
print 'NORMAL FIT SUMMARY'
print(results.summary())
print
```

###### Summary

The r squarred value is very high, however the p- value is very high as well. This is telling us that though there is some randomness
and correlation, there is some influence going on. Some of the influence is what people preceive the value of a building based on 
its location, among other factors, hipness, architecture, etc.

```python
fig = sm.qqplot(df['Estimated Gross Income'], dist=stats.distributions.norm, line='s')
plt.show()

fig = sm.qqplot(np.log(df['Estimated Gross Income']), dist=stats.distributions.norm, line='s')
plt.show()
```

```python
df_linear1 = df[['Gross Income per SqFt', 'Estimated Gross Income']]
df_linear1.columns = ['Gross_Income_per_SqFt', 'Estimated_Gross_Income']

model = smf.ols(formula='Gross_Income_per_SqFt ~ Estimated_Gross_Income' , data=df_linear1)
results = model.fit()
print 'NORMAL FIT SUMMARY'
print(results.summary())

df_linear1['LogIncome'] = np.log(df_linear1['Estimated_Gross_Income'])

log_model = smf.ols(formula='Gross_Income_per_SqFt ~ LogIncome', data=df_linear1)
log_results = log_model.fit()
print 'LOG-LOG FIT SUMMARY'
print(log_results.summary())
```

##### Summary

Changing the Estimated Gross Income improved the r squarred value, however, it is still very low.
